<h2>Imporing Libraries</h2>




```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

<h2>Reading url and data</h2>

```python
url = "https://bit.ly/4nejNue"
df = pd.read_csv(url)
```

<h1>Exploring the dataset and performing operations like mean, median</h1>


```python
print("First 5 rows of the dataset:")
print(df.head())
```

Exploring dataset

```python
print("\n Dataset Info:")
df.info()
```

```python
print("Data Types:")
print(df.dtypes)
```

```python
print("Descriptive Statistics:")
display(df.describe())
```

```python
print("Mean of columns:")
print(df.mean(numeric_only=True))
```

```python

print("Median of columns:")
print(df.median(numeric_only=True))
```

```python
print("Standard Deviation:")
print(df.std(numeric_only=True))
```

```python
print("Missing Values per Column:")
print(df.isnull().sum())
```

<h1> Visualisation </h1>

```python
print("Histograms:")
df.hist(figsize=(14, 10), bins=20, edgecolor='black', color='skyblue')
plt.suptitle("Histograms of All Numeric Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
print("Correlation Heatmap:")
plt.figure(figsize=(10, 6))
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
print("Boxplots:")
numeric_cols = df.select_dtypes(include='number').columns
plt.figure(figsize=(14, 8))
for i, col in enumerate(numeric_cols):
    plt.subplot(2, (len(numeric_cols) + 1) // 2, i + 1)
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

<h2>Value Counts for Categorical Columns</h2>

```python

print("\n Value Counts for Categorical Columns:")
categorical_cols = df.select_dtypes(include='object').columns

if len(categorical_cols) == 0:
    print(" No categorical columns found.")
else:
    for col in categorical_cols:
        print(f"\n Column: {col}")
        print(df[col].value_counts())
```
